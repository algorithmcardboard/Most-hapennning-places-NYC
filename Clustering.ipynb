```python
%matplotlib nbagg

import matplotlib.pyplot as plt
import pandas as pd
from mpl_toolkits.basemap import Basemap

df = pd.read_csv('data/uniq.dat')
X = (df.where(df['lat'].between(39.71, 41.72) & df['long'].between(-74.5,-73.5)).dropna()).as_matrix()

m = Basemap(projection='mill',llcrnrlat = 40.48,llcrnrlon = -74.26,urcrnrlat = 40.92,urcrnrlon = -73.70, resolution='h')
m.fillcontinents()

mcX, mcY = m(X[:,0], X[:,1])

land_count = 0
sea_count = 0

#print(mcX.shape[0])
print(X.shape)

for i in range(mcX.shape[0]):
    if(m.is_land(mcX[i], mcY[i])):
        land_count = land_count + 1
    else:
        sea_count = sea_count + 1
        
print(land_count, sea_count)

#plt.scatter(mcX, mcY, color='k')
```
